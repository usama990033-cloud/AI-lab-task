#### *Task(01)* **BFS without Queue & without Node**

```python
def bfsWithoutQueue (graph,startNode):
    visited=[]            # to keeping track of visited nodes
    frontier = [startNode]  # start from this node

    while frontier:
        # take the first element
        currentNode = frontier[0]
        frontier = frontier[1:]   # remove it from frontier

        if currentNode not in visited:
            print(currentNode, end=" ")   # visit the node
            visited.append(currentNode)

            # add neighbors into frontier
            for neighbor in graph[currentNode]:
                if neighbor not in visited:
                    frontier.append(neighbor)


# example graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

# calling BFS
print("BFS traversal without using queue:")
bfsWithoutQueue(graph, 'A')
```

#### *Task(02)* **BFS with Queue & Node**

```python
from collections import deque

# making a simple Node class
class Node:
    def __init__(self, value):
        self.value = value
        self.children = []   # list to store connected nodes


# BFS function using queue
def bfsWithQueue(startNode):
    visited = set()
    queue = deque([startNode])

    while queue:
        currentNode = queue.popleft()

        if currentNode not in visited:
            print(currentNode.value, end=" ")
            visited.add(currentNode)

            # add all children of current node to queue
            for child in currentNode.children:
                if child not in visited:
                    queue.append(child)


# making nodes
nodeA = Node('A')
nodeB = Node('B')
nodeC = Node('C')
nodeD = Node('D')
nodeE = Node('E')
nodeF = Node('F')

# connecting nodes (graph structure)
nodeA.children=[nodeB, nodeC]
nodeB.children=[nodeD, nodeE]
nodeC.children=[nodeF]

# calling BFS
print("BFS traversal with queue and node class:")
bfsWithQueue(nodeA)
```
