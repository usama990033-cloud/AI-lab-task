##### **MiniProject 1:**


```python
# Printing number 1 to 100

start = 1
end = 100

for i in range (start,end+1):
    # Making empty text
    word = ""

    # if number can be divided by 3,so add "Fizz"
    if i % 3 == 0:
        word = word + "Fizz"

    # if number can be divided by 5, so  add "Buzz"
    if i % 5 == 0:
        word = word + "Buzz"

    # if word is still empty, that means the number was not divisible by 3 or 5
    if word == "":
        print(i)
    else:
        print(word)
```

#### **MiniProject 2**


```python
# Mini Project 2 || Movie Budgets

movies = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]

# allowing user add more movies
extraMovies = int(input("How many extra movies? "))

for i in range(extraMovies):
    name = input("Movie name: ")
    cost = int(input("Budget: "))
    movies.append((name, cost))

# calculate total budget
totalBudget = 0
for m in movies:
    totalBudget = totalBudget + m[1]

# calculate average
movieCount = len(movies)
average = totalBudget / movieCount
print("\nAverage budget is", average)

# check which movies are above average
above = 0
print("\nMovies above average:\n")

for m in movies:
    title = m[0]
    budget = m[1]
    if budget > average:
        extra = budget - average
        print(title, "is", extra, "over average")
        above = above + 1

print("\nNumber of movies above average =", above)
```
