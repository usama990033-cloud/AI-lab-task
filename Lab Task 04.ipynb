##### **1. Code for LUHN Algorithm**

```python
#Checking the card number using the Luhn Algorithm

def luhn(cardNumber):
    # take last digit (checking remaining digit)
    checkDigit = int(cardNumber[-1])
    # reverse remaining digits of card
    digits = cardNumber[:-1][::-1]

    total = 0
    for i in range(len(digits)):
        num = int(digits[i])
        # doubling even index numbers
        if i % 2 == 0:
            num = num * 2
            if num > 9:
                num = num - 9
        total = total + num

    total = total + checkDigit
    # check divisible by 10
    return total % 10 == 0


def check_card():
    while True:
        card = input("\nEnter card number: ")
        if luhn(card):
            print(f"Card Number: {card} â†’ Valid card number âś…")
        else:
            print(f"Card Number: {card} â†’ Invalid card number âťŚ")

        # Asking user if they want do check card number again
        choice = input("\nDo you want to check another card? (y/n): ").lower()
        if choice != 'y':
            print("Exiting... Thank you!")
            break


# Run program
check_card()
```

##### **2. Write a python program to remove punctuations from the given string?**

```python
#Remove punctuation from string

while True:   
    # take input from user
    text = input("\nEnter sentence with punctuations: ")

    # make empty strings
    clean_text = ""
    removed_chars = ""

    # loop through each character
    for ch in text:
        # keeping only digits, spaces & alphabets
        if ch.isalnum() or ch.isspace():
            clean_text += ch
        else:
            removed_chars += ch

    # show results
    print("\nSentence devoid of special symbols: ", clean_text)
    if removed_chars:
        print("Characters removed:", removed_chars)
    else:
        print("No special characters were found!")

    # ask user to continue or exit (force valid input)
    while True:
        choice = input("\nWould you like to special characters to be eliminated from another text? (y/n): ").strip().lower()
        if choice in ['y', 'n']:
            break
        print("Invalid choice! Please enter 'y' for yes or 'n' for no.")

    if choice == 'n':
        print("\nGoodbye! Program exited successfully.")
        break
```

#### **3. Write a python program to sort the sentence in alphabetical order?**

```python
# Bubble sort by words & by letters
def bubbleSort(stuff):
    n = len(stuff)
    for i in range(n):
        j = 0
        while j<n-i-1:
            if str(stuff [j] ).lower()>str(stuff[j + 1]).lower():
                temp = stuff[j]
                stuff[j] = stuff[j + 1]
                stuff[j + 1] = temp
            j += 1
    return stuff

# taking input from user
sentence = input("Give input: ")

# store original
originalSentence = sentence

# Sorting by words
words = sentence.split()
bubbleSort(words)

sortedWordsSentence = ''
for word in words:
    sortedWordsSentence += word + ' '

sortedWordsSentence = sortedWordsSentence.strip()

# Sorting by letters
letters = []
for ch in sentence:
    if ch != ' ':
        letters.append(ch)

bubbleSort(letters)

sortedLettersString = ''
for l in letters:
    sortedLettersString += l

# printing everything
print("\nOriginal sentence:")
print(originalSentence)

print("\nSorted by words:")
print(sortedWordsSentence)

print("\nSorted by letters (Without spaces):")
print(sortedLettersString)
```
